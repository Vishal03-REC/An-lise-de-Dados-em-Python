```python
import pandas as pd
```

```python
candidatos = pd.read_csv('candidatos.csv',delimiter=';')
candidatos
```

```python
#CRIA UMA COLUNA ID PARA FACILITAR AS CONTAGENS
candidatos['ID']=range(1,len(candidatos)+1,1)
candidatos=candidatos[['ID','CANDIDATO','SITUACAO','DESPESAS']]
```

```python
#PERCENTUAL DE CANDIDATOS ELEITOS
# filtra apenas os candidatos eleitos e conta
eleitos = candidatos[candidatos['SITUACAO']==1]
percentual=eleitos['ID'].count()/candidatos['ID'].count()
percentual*100
```

```python
# TOP 5 DE DESPESAS
# sort_values ordena o Dataframe pelo campo DESPESAS - ordem decrescente
top_5=candidatos.sort_values(by='DESPESAS',ascending=False)
top_5.head(5)
```

```python
#MEDIA DE DESPESAS DOS ELEITOS
despesa_media_eleitos=eleitos['DESPESAS'].mean()
despesa_media_eleitos
```

```python
#QUANTOS ELEITOS GASTARAM MENOS QUE A MEDIA
eleitos_menos = eleitos[eleitos['DESPESAS']<despesa_media_eleitos]
(eleitos_menos['ID'].count()/eleitos['ID'].count())*100
```
