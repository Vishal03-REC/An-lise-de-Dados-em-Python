## Qual cargo/ função tem maior vantagem financeira em relação ao valor médio recebido em horas extras?

**Considere que o valor da coluna hora extra é o valor recebido pelo empregado em cada uma das funções assumidas durante o ano.**

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dados = pd.read_csv('salarios.csv')
```

```python
dados.columns
```

```python
dados.columns = ['Id', 'Nome', 'Funcao_Primaria', 'Funcao_Secundaria','Salario_Base', 'Valor_Horas_Extras', 'Outros_Valores',
       'Beneficios', 'Total_Pago', 'Total_Inclui_Beneficios', 'Ano', 'Nota', 'Agencia',
       'Status']
```

```python
dados.isna().sum()
```

```python
dados['Funcao_Secundaria'].fillna(dados['Funcao_Primaria'],inplace=True)
dados.isna().sum()
```

```python
dados.drop(columns = ['Status','Nota'],inplace=True)
dados.columns
```

```python
beneficios_por_funcao1 = dados['Valor_Horas_Extras'].groupby(dados['Funcao_Primaria']).mean()
beneficios_por_funcao1
```

```python
grafico1 = beneficios_por_funcao1.sort_values(ascending=False).head(15)
grafico1.plot.bar()
```

```python
beneficios_por_funcao2 = dados['Valor_Horas_Extras'].groupby(dados['Funcao_Secundaria']).mean()
```

```python
grafico2 = beneficios_por_funcao2.sort_values(ascending=False).head(15)
grafico2.plot.bar()
```

```python
beneficios_por_funcao = beneficios_por_funcao1+beneficios_por_funcao2
grafico = beneficios_por_funcao.sort_values(ascending=False).head(15)
grafico.plot.bar()
```
